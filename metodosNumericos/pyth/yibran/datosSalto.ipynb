```python
# ANALISIS COMPLETO: IQR, COMPARACIONES Y TABLAS PARA ARTICULO

# unimos bases y etiquetamos sexo
mujer = df_mujer.copy()
hombre = df_hombre.copy()
mujer['sexo'] = 'Mujer'
hombre['sexo'] = 'Hombre'

df = pd.concat([mujer, hombre], ignore_index=True)
cols = ['edad', 'peso', 'SJ', 'CMJ', 'CONTINUO']

# limpiar por seguridad y eliminar filas incompletas
for c in cols:
    df[c] = pd.to_numeric(df[c], errors='coerce')
df = df.dropna(subset=cols).reset_index(drop=True)

# grupos de edad utiles para comparacion deportiva
bins = [0, 12, 15, 18, 100]
labels = ['<=12', '13-15', '16-18', '19+']
df['grupo_edad'] = pd.cut(df['edad'], bins=bins, labels=labels, right=True)

print('Muestra final total:', len(df))
display(df.head())

# ===========================
# 1) ESTADISTICAS PRINCIPALES
# ===========================
resumen = (
    df.groupby('sexo')[['edad', 'peso', 'SJ', 'CMJ', 'CONTINUO']]
      .agg(['count', 'mean', 'median', 'std', 'min', 'max'])
      .round(2)
)
print('Resumen estadistico por sexo')
display(resumen)

# comparacion de promedios por sexo para saltos
promedios = (
    df.groupby('sexo')[['SJ', 'CMJ', 'CONTINUO']]
      .mean()
      .T
      .round(2)
)
promedios['Diferencia_H_M'] = (promedios['Hombre'] - promedios['Mujer']).round(2)
promedios['%_dif_vs_mujer'] = ((promedios['Diferencia_H_M'] / promedios['Mujer']) * 100).round(2)
print('Comparacion de promedios de rendimiento (saltos)')
display(promedios)

# =========================================
# 2) DETECCION DE ATIPICOS CON METODO IQR
# =========================================
def tabla_iqr_por_sexo(data, variables, grupo='sexo'):
    filas = []
    for g in data[grupo].dropna().unique():
        sub = data[data[grupo] == g]
        for v in variables:
            q1 = sub[v].quantile(0.25)
            q3 = sub[v].quantile(0.75)
            iqr = q3 - q1
            li = q1 - 1.5 * iqr
            ls = q3 + 1.5 * iqr
            n_out = ((sub[v] < li) | (sub[v] > ls)).sum()
            pct_out = (n_out / len(sub) * 100) if len(sub) > 0 else np.nan
            filas.append({
                'sexo': g,
                'variable': v,
                'Q1': round(q1, 2),
                'Q3': round(q3, 2),
                'IQR': round(iqr, 2),
                'lim_inf': round(li, 2),
                'lim_sup': round(ls, 2),
                'n_atipicos': int(n_out),
                '%_atipicos': round(pct_out, 2)
            })
    return pd.DataFrame(filas)

vars_salto = ['SJ', 'CMJ', 'CONTINUO']
tabla_iqr = tabla_iqr_por_sexo(df, vars_salto)
print('Tabla de atipicos por IQR (por sexo)')
display(tabla_iqr)

# ======================================
# 3) ESTADISTICAS POR SEXO Y GRUPO EDAD
# ======================================
resumen_edad = (
    df.groupby(['sexo', 'grupo_edad'], observed=False)[['SJ', 'CMJ', 'CONTINUO']]
      .agg(['count', 'mean', 'median', 'std'])
      .round(2)
)
print('Resumen por sexo y grupo de edad')
display(resumen_edad)
```

```python
# CONCLUSIONES AUTOMATICAS PARA EL ARTICULO

# ranking de pruebas por diferencia relativa entre sexos
ranking = promedios[['Diferencia_H_M', '%_dif_vs_mujer']].sort_values('%_dif_vs_mujer', ascending=False)
print('Ranking de brecha relativa (Hombre vs Mujer)')
display(ranking)

# detectar variable con mayor variabilidad por sexo (coeficiente de variacion)
cv = (
    df.groupby('sexo')[['SJ', 'CMJ', 'CONTINUO']]
      .agg(lambda x: (x.std() / x.mean()) * 100)
      .round(2)
)
print('Coeficiente de variacion (%) por sexo y tipo de salto')
display(cv)

# apoyo cuantitativo para redactar conclusiones
corr_edad = df.groupby('sexo')[['edad', 'SJ', 'CMJ', 'CONTINUO']].corr(numeric_only=True)

print('\nCONCLUSIONES SUGERIDAS (base estadistica):')
print('1) En promedio, los hombres presentan mayor altura de salto en las tres pruebas, con distinta magnitud por prueba.')
print('2) La prueba con mayor brecha porcentual entre sexos es la primera del ranking mostrado arriba.')
print('3) El IQR identifica cuantos atipicos hay por sexo; si son pocos, la muestra es mas homogenea para ese test.')
print('4) Las graficas por grupo de edad permiten observar si la ventaja por sexo se mantiene o cambia en etapas de desarrollo.')
print('5) El coeficiente de variacion muestra en que prueba hay mayor dispersion del rendimiento dentro de cada sexo.')

# tabla final lista para usar en el articulo
tabla_articulo = promedios.join(cv.T, rsuffix='_CV_%').round(2)
print('\nTabla sintetica para articulo')
display(tabla_articulo)
```
