```python
# haremos un grafico de caja y un histograma por saltos (1. SJ, 2. CMJ, 3. CONTINUO)

# HISTOGRAMAS
for col in ['SJ', 'CMJ', 'CONTINUO']:
    plt.figure(figsize=(8,4))   # siempre se me olvida hacer el recuadro eghhhh 
    sns.histplot(df_mujer[col], color='pinK', stat='count', label='Mujeres', kde=True)
    sns.histplot(df_hombre[col], color='blue', kde=True, label='Hombre', stat='count')  # count= num saltos, density es la densidad (proporcion por unidad de saltos)
    plt.title(f"HISTOGRAMA DE SALO {col}")
    plt.legend()
    plt.show()
```

```python
# BOXPLOTS
for col in ['SJ', 'CMJ', 'CONTINUO']:
    datos = pd.DataFrame({
        'valor': pd.concat([df_mujer[col], df_hombre[col]]),
        'sexo': ['Mujer']*len(df_mujer) + ['Hombre']*len(df_hombre)
    })
    plt.figure(figsize=(8,4))
    sns.boxplot(x='sexo', y='valor', data=datos, palette='Set2')
    plt.title(f'Diagrama de cajas (H vs M) del salto {col}')
    plt.legend()
    plt.show()  
```
