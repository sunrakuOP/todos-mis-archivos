```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# OPCIONAL: Estilo bonito para las gráficas
sns.set_theme(style="whitegrid")

# === 1. CARGAR LOS DATOS ORIGINALES ===
df = pd.read_csv("datos_taller.csv")  # Cambia el nombre de archivo si es diferente

print("\nPrimeras filas del archivo:")
print(df.head())
print("\nTipos de datos:")
print(df.info())
print("\nDatos nulos por columna:")
print(df.isnull().sum())

# === 2. LIMPIEZA DE DATOS ===
# Ejemplo de conversión de columna a numérica
# Cambia 'columna1', 'columna2' por los nombres correctos reales de tu archivo
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')  # Solo convierte si es posible

# Ejemplo eliminando outliers y valores imposibles
if 'temperatura' in df.columns:
    df.loc[df['temperatura'] > 60, 'temperatura'] = np.nan

# === 3. RELLENAR DATOS FALTANTES ===
for col in df.columns:
    if df[col].dtype in [np.float64, np.int64]:
        df[col] = df[col].interpolate()

print("\nDespués de limpiar:")
print(df.info())

# === 4. GUARDAR RESULTADOS LIMPIOS ===
df.to_csv('datos_taller_limpio.csv', index=False)
print("\nArchivo limpio guardado como datos_taller_limpio.csv")

# === 5. GRAFICAR ALGUNA VARIABLE (OPCIONAL) ===
if 'x' in df.columns and 'y' in df.columns:
    plt.figure(figsize=(8, 5))
    plt.plot(df['x'], df['y'], label='Serie limpia')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.title('Ejemplo gráfico de x vs y')
    plt.show()
```
