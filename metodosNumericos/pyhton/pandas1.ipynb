```python
import pandas as pd   # importamos una libreria pandas 

#crear un diccionario, metemos datos crudos manualmente
data = {
    "nombre": ["ana", "juan", "pedro"],
    "edad": [23, 35, 28],
    "ciudad": ["madrid", "bogota", "cdmx"],
    'fecha': ['01/01/2024 09:00', '2024-01-01 10:00:00', 'Error_Date']
}

df = pd.DataFrame(data)    # variable dataframe que almacena la tabla de datos
# pd.DataFrame(datos)  funcion que crea una tabla de datos
print(df)

df.to_csv('primeroPrueba.csv', index = False)    
df1 = pd.read_csv('primeroPrueba.csv')
display(df1)  
df1.info()

# LIMPIEZA
    #1. LIMPIEZA DE FEHAS
# cojo columna fechas y la limpio de errores entregando objetos tipo DATETIME
df1['fecha'] = pd.to_datetime(df1['fecha'], errors = "coerce")
display(df1)  # para mostrar fecha corregida

    #2. LIMPIEZA DE DATOS
# convertir numeros en texto 
df1['edad'] = pd.to_numeric(df1['edad'], errors = "coerce")
display(df1)
```
